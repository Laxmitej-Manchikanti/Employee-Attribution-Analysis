```python
plt.figure(figsize=(10,6))
sns.histplot(data['Age'],kde=True)
plt.title("Distribution of Age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Select categorical columns
categorical_columns = data.select_dtypes(include=['object']).columns

# Apply LabelEncoder to each categorical column
for column in categorical_columns:
    data[column] = le.fit_transform(data[column])

# Display the first few rows of the encoded dataframe
data.head()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix', fontsize=18)
plt.show()
```

```python
# Count of attrition
attrition_count = data['Attrition'].value_counts()
print("Attrition count:\n", attrition_count)

# Plotting attrition count
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data = data)
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Attrition', y='Age', data=data)
plt.title('Attrition vs Age')
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='JobRole', hue='Attrition', data=data)
plt.title('Attrition by Job Role')
plt.xticks(rotation=90)
plt.show()
```

```python
cleaned_file_path = '/content/Cleaned_Attrition_data.csv'
data.to_csv(cleaned_file_path, index=False)
print(f"Cleaned dataset saved to {cleaned_file_path}")
```
